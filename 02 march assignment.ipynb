Q2)

Scatter plot is a tool used to visualise data under matplotlib.


```python
import numpy as np
np.random.seed(3)
x=3+np.random.normal(0,2,50)
y=3+np.random.normal(0,2,len(x))
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(x,y)
plt.xlabel("speed")
plt.ylabel("death")
plt.title("scatterplot")
```

Bar plot is a tool in python Matplotlib library used to create bar graphs for visulisation.


```python
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
```
